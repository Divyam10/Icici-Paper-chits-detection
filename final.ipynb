```python
import cv2 as cv
import numpy as np
import tensorflow as tf




font = cv.FONT_HERSHEY_SIMPLEX 
# Read the graph.
with tf.gfile.FastGFile('frozen_inference_graph.pb', 'rb') as f:
    graph_def = tf.GraphDef()
    graph_def.ParseFromString(f.read())

with tf.Session() as sess:
    # Restore session
    sess.graph.as_default()
    tf.import_graph_def(graph_def, name='')

    # Read and preprocess an image.
    img = cv.imread('new2/t (31).jpg')
    rows = img.shape[0]
    cols = img.shape[1]
    inp = cv.resize(img, (300, 300))
    inp = inp[:, :, [2, 1, 0]]  # BGR2RGB

    # Run the model
    out = sess.run([sess.graph.get_tensor_by_name('num_detections:0'),
                    sess.graph.get_tensor_by_name('detection_scores:0'),
                    sess.graph.get_tensor_by_name('detection_boxes:0'),
                    sess.graph.get_tensor_by_name('detection_classes:0')],
                   feed_dict={'image_tensor:0': inp.reshape(1, inp.shape[0], inp.shape[1], 3)})
    
    
    
    # Visualize detected bounding boxes.
    num_detections = int(out[0][0])
    person=False
    for i in range(num_detections):
        
        
        
        classId = int(out[3][0][i])
       
        if classId!=1:
            
            continue
       
        else:
            person=True
            score = float(out[1][0][i])
            bbox = [float(v) for v in out[2][0][i]]
            if score > 0.3:
                x = bbox[1] * cols
                y = bbox[0] * rows
                right = bbox[3] * cols
                bottom = bbox[2] * rows
                cv.rectangle(img, (int(x), int(y)), (int(right), int(bottom)), (125, 255, 51), thickness=2)

    if person==True: 
        img = cv.putText(img,'Person is present',(23,370), font, 1,(169, 245, 159),4,cv.LINE_AA)

        cv.imshow('TensorFlow MobileNet-SSD', img)
        cv.waitKey(0)
        cv.destroyAllWindows()

    else:
        image = cv.imread('new2/t (31).jpg')
        image = cv.GaussianBlur(image,(5,5),0)
        img = image[249:375,202:394]


        kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
        dilation = cv.dilate(img,kernel,iterations = 3)

        dilation = cv.dilate(dilation, None, iterations=1)
        dilation = cv.erode(dilation, None, iterations=1)



        img1 = dilation.copy()
        for i in range(np.shape(dilation)[0]):
            for j in range(np.shape(dilation)[1]):
                b,g,r = dilation[i][j]

                l= 80
                u= 130

                if l<b<u and l<g<u and l<r<u:
                    b,g,r      = 255,255,255
                    img1[i][j] =  b,g,r
                else :
                    b,g,r      = 0,0,0
                    img1[i][j] =  b,g,r




        img2 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
        contours, hierarchy = cv.findContours(img2,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)

        pointer=0
        for i in range(len(contours)):
            a=cv.contourArea(contours[i])
            if a>0 and a<600:
                    cv.drawContours(img, contours, i, (0,255,0), 1)
                    cv.drawContours(dilation, contours, i, (0,255,0), 1)
                    pointer+=1


        font = cv.FONT_HERSHEY_SIMPLEX 
        if pointer>1:
            image = cv.putText(image,'May not be clean',(23,370), font, 1,(12, 12, 232),4,cv.LINE_AA)
        else:
            image = cv.putText(image,'may be clean',(23,370), font, 1,(169, 245, 159),4,cv.LINE_AA)



        #print(k)
        cv.imshow('image',image)
        #cv.imshow('cropped',img)
        #cv.imshow('threshold', img1)
        #cv.imshow('dilation', dilation)
        cv.waitKey(0)
        cv.destroyAllWindows()    
```
